## Setup

### Imports

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import rcParams
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from matplotlib.legend_handler import HandlerTuple
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

import scipy.integrate as integrate

# matplotlib style sheet
plt.style.use('../style_1.mplstyle')
```

### Lines, colours, legends

```python
# Line style list
lineStyle41 = ['-', ':', '--', '-.']

# Colours from colour map "tab10"; blue, red, green, orange, purple
colour_map_colours = [plt.cm.tab10(i) for i in [0, 3, 2, 1, 4]]

# Add black as first colour
colourList = ['k'] + colour_map_colours

# Labels for lines
compLabels = ['HFB', 'ETFSI', 'ETFSI + pairing']

# Position and formatting of legend
top_legend_kwargs = dict(
    loc='lower left',
    bbox_to_anchor=(0, 1.05, 1, 0.5),
    ncol=3,
    borderpad=0.5,
    labelspacing=0.,
    handlelength=2.,
    handletextpad=0.4,
    borderaxespad=0.,
    edgecolor='k',
    mode='expand',
    fancybox=False)

# Text box properties
text_bbox = dict(boxstyle='round',
                 facecolor=(1, 1, 1, 1),
                 edgecolor='k',
                 pad=0.5)

# Letter labels
panelLabels = ['a', 'b', 'c', 'd']
```

# ETF vs. HFB: multiple slices (SLy4 and BSk24)

## Load data

```python
## SLy4 ##
SLy4_hfb_slices = []
SLy4_hfb_slices.append(np.loadtxt('data/SLy4/hfb/slice_0.00041243.dat'))
SLy4_hfb_slices.append(np.loadtxt('data/SLy4/hfb/slice_0.0013.dat'))
SLy4_hfb_slices.append(np.loadtxt('data/SLy4/hfb/slice_0.005.dat'))

SLy4_noPair_slices = []
SLy4_noPair_slices.append(np.loadtxt('data/SLy4/etf_noPair/all_energies_rho_b_0.000412435199.dat'))
SLy4_noPair_slices.append(np.loadtxt('data/SLy4/etf_noPair/all_energies_rho_b_0.0013.dat'))
SLy4_noPair_slices.append(np.loadtxt('data/SLy4/etf_noPair/all_energies_rho_b_0.005.dat'))

SLy4_pair_slices = []
SLy4_pair_slices.append(np.loadtxt('data/SLy4/etf_pair/all_energies_rho_b_0.000412435199_20_60.dat'))
SLy4_pair_slices.append(np.loadtxt('data/SLy4/etf_pair/all_energies_rho_b_0.0013_20_60.dat'))
SLy4_pair_slices.append(np.loadtxt('data/SLy4/etf_pair/all_energies_rho_b_0.00500000_20_60.dat'))

# Adjust HFB data
for i, hfb in enumerate(SLy4_hfb_slices):
    # Number of particles
    A = hfb[:,0] * 4./3 * np.pi * hfb[:,3]**3
    
    # Add on ETF kinetic and potential energies of electrons, and Q_beta factor
    SLy4_hfb_slices[i] = np.column_stack(
        (hfb, hfb[:,2] + (np.sum(SLy4_pair_slices[i][:,15:17], axis=1) - 0.782343*hfb[:,1]) / A))


## BSk24 ##
BSk24_hfb_slices = []
BSk24_hfb_slices.append(np.loadtxt('data/BSk24/hfb/slice_0.00041243.dat'))
BSk24_hfb_slices.append(np.loadtxt('data/BSk24/hfb/slice_0.0013.dat'))
BSk24_hfb_slices.append(np.loadtxt('data/BSk24/hfb/slice_0.005.dat'))

BSk24_noPair_slices = []
BSk24_noPair_slices.append(np.loadtxt('data/BSk24/noPair_slices/all_energies_rho_b_0.000412435199_20_60.dat'))
BSk24_noPair_slices.append(np.loadtxt('data/BSk24/noPair_slices/all_energies_rho_b_0.0013_20_60.dat'))
BSk24_noPair_slices.append(np.loadtxt('data/BSk24/noPair_slices/all_energies_rho_b_0.00500000.dat'))

BSk24_pair_slices = []
BSk24_pair_slices.append(np.loadtxt('data/BSk24/pair_slices/all_energies_rho_b_0.000412435199_20_60.dat'))
BSk24_pair_slices.append(np.loadtxt('data/BSk24/pair_slices/all_energies_rho_b_0.0013_20_60.dat'))
BSk24_pair_slices.append(np.loadtxt('data/BSk24/pair_slices/all_energies_rho_b_0.00500000_20_60.dat'))

# Adjust HFB data
for i, hfb in enumerate(BSk24_hfb_slices):
    A = hfb[:,0] * 4./3 * np.pi * hfb[:,3]**3
    
    BSk24_hfb_slices[i] = np.column_stack(
        (hfb, hfb[:,2] + (np.sum(BSk24_pair_slices[i][:,15:17], axis=1) - 0.782343*hfb[:,1]) / A))
```

## Plot

```python
# Labels and their positions for each slice subplot
subplotLabels = [r'$\rho_b = 0.00041$', r'$\rho_b = 0.0013$', r'$\rho_b = 0.005$']
xLoc = [40, 40, 40]
yLoc = [-0.85, 0.465, 2.025]

fig, axs = plt.subplots(3, 1, sharex=True, figsize=(3.4, 4.5), gridspec_kw={'hspace': 0})

for i, ax in enumerate(axs):
    ax.plot(SLy4_hfb_slices[i][:,1], SLy4_hfb_slices[i][:,4],
            linestyle=lineStyle41[0], color=colourList[0], label=compLabels[0])
    
    ax.plot(SLy4_noPair_slices[i][:,1], SLy4_noPair_slices[i][:,23],
            linestyle=lineStyle41[1], color=colourList[1], label=compLabels[1])
    ax.plot(SLy4_pair_slices[i][:,1], SLy4_pair_slices[i][:,23],
            linestyle=lineStyle41[2], color=colourList[2], label=compLabels[2])
    
    ax.set_xlim([20,60])
    ax.set(xlabel=r'$Z$', ylabel=r'$e \left[\mathrm{MeV}\right]$')
    
    # Ticks and grid
    ax.xaxis.set_major_locator(MultipleLocator(10))
    ax.xaxis.set_minor_locator(MultipleLocator(2))
    
    ax.text(xLoc[i], yLoc[i],
            subplotLabels[i],
            ha='center', va='center',
            bbox = text_bbox)

# Create legend at top of figure
legend = axs[0].legend(**top_legend_kwargs)

# Align labels of y-axes
fig.align_ylabels(axs[:])

fig.tight_layout()
fig.savefig('SLy4_slices.pdf')
```

## Plot SLy4 and BSk24 together

```python
# Labels and their positions for each slice subplot
subplotLabels = [r'$\rho_b = 0.00041$', r'$\rho_b = 0.0013$', r'$\rho_b = 0.005$']
xLoc = [40, 40, 40, 40, 40, 40]
yLoc = [-0.85, 0.46, 2.05, -0.85, 0.46, 2.05]

# y-axis limits
yLims = [[-1.11, -0.78], [0.24, 0.49], [1.89, 2.16]]

fig, axs = plt.subplots(3, 2, sharex=True, figsize=(7., 4.5), gridspec_kw={'hspace': 0})

for i in range(3):
    # SLy4
    axs[i,0].plot(SLy4_hfb_slices[i][:,1], SLy4_hfb_slices[i][:,4],
            linestyle=lineStyle41[0], color=colourList[0], label=compLabels[0])
    axs[i,0].plot(SLy4_noPair_slices[i][:,1], SLy4_noPair_slices[i][:,23],
            linestyle=lineStyle41[1], color=colourList[1], label=compLabels[1])
    axs[i,0].plot(SLy4_pair_slices[i][:,1], SLy4_pair_slices[i][:,23],
            linestyle=lineStyle41[2], color=colourList[2], label=compLabels[2])
    
    # BSk24
    axs[i,1].plot(BSk24_hfb_slices[i][:,1], BSk24_hfb_slices[i][:,4],
            linestyle=lineStyle41[0], color=colourList[0], label=compLabels[0])
    axs[i,1].plot(BSk24_noPair_slices[i][:,1], BSk24_noPair_slices[i][:,23],
            linestyle=lineStyle41[1], color=colourList[1], label=compLabels[1])
    axs[i,1].plot(BSk24_pair_slices[i][:,1], BSk24_pair_slices[i][:,23],
            linestyle=lineStyle41[2], color=colourList[2], label=compLabels[2])
    
    axs[i,0].set_ylabel(r'$e \left[\mathrm{MeV}\right]$')
    axs[i,1].set_ylabel(r'$e \left[\mathrm{MeV}\right]$')
    
    # y-axis limits
    axs[i,0].set_ylim(yLims[i])
    axs[i,1].set_ylim(yLims[i])
    
# Column titles
for ax, col in zip(axs[0,:], ['SLy4', 'BSk24']):
    ax.set_title(col)

for i, ax in enumerate(np.ravel(axs.T)):
    # x-axis limits
    ax.set_xlim([20, 60])
    
    # Ticks and grid
    ax.xaxis.set_major_locator(MultipleLocator(10))
    ax.xaxis.set_minor_locator(MultipleLocator(2))
    #ax.yaxis.set_major_locator(MultipleLocator(0.05))
    
    # Text boxes
    ax.text(xLoc[i], yLoc[i],
                      subplotLabels[i%3],
                      ha='center', va='center',
                      bbox=text_bbox)

for i, ax in enumerate(axs[-1,:]):
    ax.set_xlabel(r'$Z$')

# y-axis labels for BSk24 on right
for ax in axs[:,1]:
    ax.yaxis.tick_right()
    ax.yaxis.set_label_position("right")

# Position and formatting of legend
SLy4_BSk24_legend_kwargs = dict(
    loc='upper center',
    bbox_to_anchor=(0., 0., 1., 0.99),
    ncol=3,
    borderpad=0.5,
    columnspacing=1.,
    handlelength=2.,
    handletextpad=0.35,
    borderaxespad=0.,
    edgecolor='k',
    fancybox=False)

# Grouping of lines in legend entries
lines = axs[0,0].get_lines()

# Legend
SLy4_BSk24_legend = fig.legend(lines, compLabels, **SLy4_BSk24_legend_kwargs)

# Align labels of y-axes
fig.align_ylabels(axs[:,0])
fig.align_ylabels(axs[:,1])

fig.tight_layout()
fig.savefig('slices_SLy4_BSk24.pdf')    
```

# Accuracy of LDA for neutron pairing

## Functions for energies and energy densities

```python
# Single-particle kinetic energy (relative to effective chemical potential)
def xi_k(k, mu):
    return k**2/2. - mu

# Quasiparticle energy
def E_k(k, mu, Delta):
    return np.sqrt(xi_k(k, mu)**2 + Delta**2)


# Integrand for energy density (pairing + kinetic)
def fepsilon(k, mu, Delta):
    return k**2/(2.*np.pi**2) * (-Delta**2/(2.*E_k(k,mu,Delta)) + k**2/2. * (1.-xi_k(k,mu)/E_k(k,mu,Delta)))

# Corresponding integral from 0 to Lambda
def epsilon(mu, Delta):
    integral, err = integrate.quad(fepsilon, 0, Lambda, (mu, Delta))
    return integral


# Integrand for density
def frho(k, mu, Delta):
    return (k**2 / (2.*np.pi**2) * (1. - xi_k(k, mu) / E_k(k, mu, Delta)))

# Corresponding integral from 0 to Lambda
def rho(mu, Delta):
    integral, err = integrate.quad(frho, 0, Lambda, (mu, Delta))
    return integral


# Fermi momentum
def k_F(mu, Delta):
    return ((3.*np.pi**2 * rho(mu, Delta))**(1./3))

# Fermi energy
def e_F(mu, Delta):
    return k_F(mu, Delta)**2 / 2.


# Kinetic energy density of normal (unpaired) state with same density
def epsilon_0(mu, Delta):
    return k_F(mu, Delta)**5 / (10.*np.pi**2)
```

## Gains in energy-per-particle $e$

```python
# True energy gain per particle
def del_epsilon(mu, Delta):
    return (epsilon(mu, Delta) - epsilon_0(mu, Delta)) / rho(mu, Delta)

# Weak-coupling result for energy gain per particle
def del_epsilon_weak(mu, Delta):
    return -3./8 * (Delta**2/mu)

# Broglia's formula for energy gain per particle
def del_epsilon_Broglia(mu, Delta):
    return -3./8 * (Delta**2/e_F(mu, Delta))


# Cut-off momentum
Lambda = 15.

# True energy gain per particle
Del_grid = np.linspace(0.03, 0.3, 100)
normal = np.empty(100)
for i in range(0, 100):
    normal[i] = del_epsilon(1., Del_grid[i])

# Precision of weak-coupling result as function of Delta/mu, with mu=1
weak = del_epsilon_weak(np.ones(100), Del_grid)

# Precision of Broglia's formula as function of Delta/mu, with mu=1
Broglia = np.empty(100)
for i in range(0, 100):
    Broglia[i] = del_epsilon_Broglia(1., Del_grid[i])

fig, ax = plt.subplots(figsize=(3.4, 3))
ax.plot(Del_grid, normal/weak,
        linestyle=lineStyle41[0], color=colourList[0], label=r'Using $\mu$: $-\frac{3\Delta^2}{8\mu}$')
ax.plot(Del_grid, normal/Broglia,
        linestyle=lineStyle41[1], color=colourList[1], label=r'Using $\varepsilon_F$: $-\frac{3\Delta^2}{8\epsilon_F}$')

ax.set_xlim([0., 0.3])
ax.set_ylim([1., 1.1])

ax.set_xlabel(r'$\Delta/\mu$')
ax.set_ylabel(r'$\delta e/\delta e_{weak}$')

ax.xaxis.set_major_formatter(FormatStrFormatter('%g')) # No unnecessary trailing decimals
ax.grid(True, which='both')
ax.legend()

fig.tight_layout()
fig.savefig('weak-coupling_accuracy.pdf')
```
